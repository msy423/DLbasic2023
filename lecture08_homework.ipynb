**Q2 CAMとGrad-CAMの説明として最も「適切でない」ものを選択せよ**
1. CAMは，Global Average Poolingの直後に出力全結合層がある形のモデルにしか適用できない．
1. Grad-CAMは，特徴マップの重みづけに勾配を用いるので，微分可能であれば畳み込み以外のレイヤーにも適用することができる．
1. Grad-CAMをGuided Backpropagationと組み合わせることで，モデルの判断根拠をより高解像度で表示できる．
1. Grad-CAMは，モデルの判断だけでなく，任意のクラスについて可視化することができる．

回答: $\fbox{　(1)　}$

**Q3 SHAP (Shapley Additive Explanations) の説明として最も「適切でない」ものを選択せよ**
1. 特徴量の数を$n$としたとき，理論的にはShapley valueの計算に$2^n$回の推論が必要となるが，SHAPでは線形回帰や決定木を用いてこれを近似する．
1. ゲーム理論におけるShapley valueとは，複数のプレイヤーが共同で働き一つの結果を生み出すとき，それぞれのプレイヤーの結果への寄与率を計算するものである．
1. ある特徴量のSHAP値は，元のモデル出力と，その特徴量だけを取り除いたときのモデル出力の差によって計算される．
1. モデルの出力に影響を与えない特徴量のSHAP値はゼロになる．

回答: $\fbox{　(2)　}$
